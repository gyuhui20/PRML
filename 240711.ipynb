## 2.1 Binary Variables

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
```

```python
#Bernoulli dist.
def bern(mu, x):
    return (mu**x) * (1-mu)**(1-x)
#print(bern(0.4,1))
```

```python
mu = 0.4
datas_bern = []
for i in range(1000):
    prob = np.random.rand(1)
    if prob <= mu:
        datas_bern.append(1)
    else :
       datas_bern.append(0)
print(datas_bern)

plt.hist(datas_bern)
plt.title('Bernoulli dist. from data')
plt.xlabel('x')
plt.ylabel('number(count)')
plt.show()
```

```python
def log_likelihood_bern(mu, data_list):
    result = 0
    for data in data_list:
        result += np.log(bern(mu, data))
    return result
```

```python
mu_list = [0.01*i for i in range(101)]
likelihood_list = [log_likelihood_bern(i, datas_bern) for i in mu_list]
print(likelihood_list)

plt.plot(mu_list,likelihood_list)
plt.title(f'Maximum log likelihood: %s with mu = %s' %(f'{np.max(likelihood_list):.2f}',f'{mu_list[np.argmax(likelihood_list)]:.1f}'))
plt.xlabel('mu')
plt.show()
```

```python
N = 10
mu = 0.4
```

```python
def binomial(m, N, mu):
    return sp.special.comb(N,m)*mu**m*(1-mu)**(N-m)

m_list = [m for m in range(N+1)]
p_list = [binomial(m,N,mu) for m in m_list]
```

```python
plt.bar(m_list, p_list)
plt.title('Binomial dist. from function')
plt.xlabel('x')
plt.ylabel('number(count)')
plt.show()
```

```python
datas_bin = np.random.binomial(N, mu, 10000)
print(datas_bin)

def log_likelihood_bin(data, N, mu):
    result = 0
    for i_data in data:
        prob = binomial(i_data, N, mu)
        result += np.log(prob)
    return result
```

```python
mu_list = [0.01*i for i in range(101)]
log_likelihoods_bin = [log_likelihood_bin(datas_bin, N, mu) for mu in mu_list]

print(mu_list)
print(log_likelihoods_bin)

plt.plot(mu_list, log_likelihoods_bin)
plt.title(f"Maximum log likelihood: {np.max(log_likelihoods_bin):.2f} with mu: {mu_list[np.argmax(log_likelihoods_bin)]}")
plt.xlabel('mu')
plt.show()
```
